## 콜라츠 추측

image.png

```python
def solution(n):
    count = 1
    while count < 500:
        if n % 2 == 0 :
            n = n/2
            if n == 1:
                return count
                break
        else:
            n = n*3 + 1
            if n == 1:
                return count
                break
        count += 1
    return -1
# 테스트 13 실패
```

```python
# 테스트 13 실패, n=1인경우 추가
def solution(n):
    count = 1
    if n == 1:
        return 0
    while count < 500:
        if n % 2 == 0 :
            n = n/2
            if n == 1:
                return count
                break
        else:
            n = n*3 + 1
            if n == 1:
                return count
                break
        count += 1
    return -1
```

## 서울에서 김서방 찾기

image.png

```python
def solution(seoul):
    for i,j in enumerate(seoul):
        if j == 'Kim':
            return f"김서방은 {i}에 있다"
```

```python
def solution(seoul):
    return [f"김서방은 {i}에 있다" for i,j in enumerate(seoul) if j == 'Kim'][0]
```

```python
def solution(seoul):
    return "김서방은 {}에 있다".format(seoul.index("Kim"))
```
