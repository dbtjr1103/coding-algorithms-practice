```python
def solution(s):
    count, d_zero = 0,0
    while s != "1":
        d_s = len(s)
        s = s.replace('0','')
        d_zero += d_s - len(s)
        count += 1
        s = bin(len(s))[2:]
    return [count, d_zero]
```

```python
solution("110010101001")
```
