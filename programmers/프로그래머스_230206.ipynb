## 나누어 떨어지는 숫자 배열

image.png

```python
# 일단 풀어보기
def solution(arr, divisor):
    answer = []
    for i in arr:
        if i % divisor ==0:
            answer.append(i)
    if answer == []:
        return [-1]
    return sorted(answer)
```

```python
# 정리
def solution(arr, divisor):
    answer = [i for i in arr if i % divisor == 0]
    return sorted(answer) if answer != [] else [-1]
```
