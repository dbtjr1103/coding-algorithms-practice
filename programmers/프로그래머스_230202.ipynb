## 나머지가 1이 되는 수 찾기

image.png

```python
def solution(n):
    x = 2
    while n % x != 1:
        x += 1
    return x
```

```python
def solution(n):
    return [x for x in range(1,n+1) if n%x==1][0]
```

```python
def solution(n):
    return next(i for i in range(2, n) if n%i == 1)
solution(3)
```

## 두 정수 사이의 합

image.png

```python
def solution(a, b):
    if a<b:
        return sum(range(a, b+1))
    return sum(range(b, a+1))
```
