```python
n = 78
# bin(n)
str(bin(n)).count('1')
```

```python
def solution(n):
    i = str(bin(n)).count('1')
    while 1:
        n += 1
        if str(bin(n)).count('1') == i:
            break
    return n
```

```python
solution(n)
```
