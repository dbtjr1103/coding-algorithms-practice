```python
people = [70, 50, 80, 50]
limit = 100
```

```python
def solution(people, limit):
    people.sort()
    answer = 0
    
    while len(people)>=2:
        if people[0] + people[-1] <= limit:
            del people[0], people[-1]
            answer += 1
        else:
            del people[-1]
            answer += 1
            
    if people:
        return answer + 1
        
    return answer
```

```python
def solution(people, limit) :
    answer = 0
    people.sort()

    a = 0
    b = len(people) - 1
    while a < b :
        if people[b] + people[a] <= limit :
            a += 1
            answer += 1
        b -= 1
    return len(people) - answer
```
