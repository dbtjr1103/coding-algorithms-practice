## 가운데 글자 가져오기

image.png

```python
def solution(s):
    ls = len(s)
    if ls % 2 == 0:
        return s[int(ls/2-1):int(ls/2+1)]
    else:
        return s[int(ls/2)]
s = 'abcde'
solution(s)
```
