## 정수 내림차순으로 배치하기

image.png

```python
def solution(n):
    return int("".join(sorted(str(n), reverse=True)))
```

## 하샤드 수

```python
def solution(x):
    a = sum(int(i) for i in str(x))
    if x % a == 0:
        return True
    return False
```
