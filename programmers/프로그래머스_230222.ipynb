## 크기가 작은 부분문자열

image.png

```python
def solution(t, p):
    answer = 0
    p_len, t_len = len(p), len(t)
    for i in range(t_len-p_len+1):
        if int(t[i:i+p_len]) <= int(p):
            answer += 1
    return answer
```

```python
def solution(s):
    answer = []
    for i in range(len(s)):
        n = -1
        for j in range(i-1, -1, -1):
            if s[i] == s[j]:
                n = i - j
                break
        answer.append(n)
    return answer
```

```python
def solution(s):
    answer = []
    dic = dict()
    for i in range(len(s)):
        if s[i] not in dic:
            answer.append(-1)
        else:
            answer.append(i - dic[s[i]])
        dic[s[i]] = i

    return answer
```

```python
def solution(k, score):
    answer = []
    rank = []
    for i in score:
        rank.append(i)
        rank.sort(reverse=True)
        if len(rank)>k:
            del rank[-1]
        answer.append(rank[-1])
    return answer
```
