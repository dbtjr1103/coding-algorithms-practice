문자열 내 p와 y의 개수

image.png

```python
import collections

def solution(s):
    s = s.lower()
    tmp = collections.Counter(s)
    
    if tmp['p'] == tmp['y']:
        return True
    return False
```

```python
solution("pPoooyY")
```

```python
def solution(s):
    return s.lower().count('p') == s.lower().count('y')
```

```python
solution("pPoooyY")
```

### x만큼 간격이 있는 n개의 숫자

image.png

```python
def solution(x, n):
    answer = list(i for i in range(x, x*n, x))
    return answer + [x*n]
```
