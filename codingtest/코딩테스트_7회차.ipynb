## 3. 이상한 문자 만들기

image.png

```python
import time 

start = time.time()

def solution(s):
# 답안 작성 부분 ===============
    answer = ''
    for i in range(len(s)):
        if i%2==0:
            answer += s[i].upper()
        else:
            answer += s[i].lower()

# ==============================

    return answer

print(solution("try hello world"))
print(solution("Big one is  coming"))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

```python
# 그리디 알고리즘
"""탐욕법 - 현재 상황에서 지금 당장 좋은 것만 고르는 방법 : 최적의 해를 보장할 수 없을때가 많다.
정당성 분석이 매우 중요, 단순히 가장 좋아보이는 것을 반복적으로 선택해도 최적의 해를 구할 수 있는지 검토
코딩테스트에서는 탐욕법으로 얻은 해가 최적의 해가 되는 상황에서 문제를 출제함"""

# 1이 될 때까지
"""어떠한 수 n이 1이 될 때까지 두 과정중 하나를 반복저긍로 수행
 1. N에서 1을 뺍니다.
 2. N을 K로 나눕니다. 
  => N에 대하여 최대한 많이 나누기를 수행 
      -> 2이상의 수로 나누는 작업이 1을 빼는 작업보다 훨씬 많이 줄일 수 있다"""

# 문제풀이
n, k = map(int, input().split()) # N, K를 공백을 기준으로 구분하여 입력받기

result = 0 # 연산 수행 횟수

while True:
    target = (n // k) * k # n이 k로 나누어 떨어지지 않을때 가장 가까운 n으로 떨어지는 수를 찾을 수 있음
    result += (n-target) # 1을 빼는 연산횟수를 한번에 더해줌 
    n = target

    if n<k: #n이 k보다 작을 때 반복문 탈출
        break
    
    result += 1
    n //= k # K로 나누기 1회

result += (n-1) #나누지 못한 나머지를 1까지 빼줌
print(result)
```
