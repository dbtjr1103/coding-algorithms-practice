## 1. 피보나치 수

image.png

```python
import time 

start = time.time()

def solution(n):
# 답안 작성 부분 ===============
    answer = [1,1]
    for i in range(2, n):
        answer.append((answer[-1]+answer[-2]) % 1234567 )
        
    return answer[-1] % 1234567
# ==============================

print(solution(3))
print(solution(5))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

## 2. 짝지어 제거하기

image.png

```python
import time 

start = time.time()

def solution(s):
# 답안 작성 부분 ===============
    stack = []
    for i in s:
        if stack and stack[-1] == i:
            stack.pop()
        else:
            stack.append(i)
    if stack:
        return 0
    return 1
# ==============================

    # return answer

print(solution("baabaa"))
print(solution("cdcd"))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

```python
import time 

start = time.time()

def solution(s):
# 답안 작성 부분 ===============

    count, zero = 0, 0

    while s != '1':
        count += 1
        zero += s.count('0')
        s = bin(s.count('1'))[2:]
    return [count, zero]

# ==============================


print(solution("110010101001"))
print(solution("01110"))
print(solution("1111111"))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```
